## Question 8: Find the top 5 bestselling pizzas.

```python
# Find the top 5 bestselling pizzas
top_5_pizzas = (order_list.groupby("name")["quantity"].sum().sort_values(ascending=False).head(5))

print(top_5_pizzas)
```

```python
# Plot the chart

top_5_pizzas.plot(kind="bar")

plt.title("Top 5 Bestselling Pizzas")
plt.xlabel("Pizza Name")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)

plt.show()
```

## Question 8: Find the sales made in each month. Any trend noticeable?

```python
# Create a month column
order_list["month"] = order_list["date"].dt.month_name()

# Calculate total sales for each month
monthly_sales = order_list.groupby("month")["sales"].sum()

# Arrange months in the correct order
month_order = ["January", "February", "March", "April","May", "June", "July", "August","September", "October", "November", "December"]

monthly_sales = monthly_sales.reindex(month_order)

print(monthly_sales)
```

```python
# Find the month with the highest sales

highest_sales_month = monthly_sales.idxmax()
highest_monthly_sales = monthly_sales.max()

print("Month with Highest Sales:", highest_sales_month)
print("Highest Monthly Sales:", highest_monthly_sales)
```

### Plot the monthly sales trend using line chart

```python
monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```
